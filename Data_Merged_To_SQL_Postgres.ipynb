```python
import pandas as pd
```

```python
from sqlalchemy import create_engine
```

```python
url_1="datasets/costs_College_Data_777.csv"
```

```python
costs_college=pd.read_csv(url_1)
```

```python
costs_college.head()
```

```python
url_2="datasets/salary_potential.csv"
```

```python
salary=pd.read_csv(url_2)
```

```python
salary.head()
```

```python
salary.name.count()
```

```python
merge1=costs_college.merge(salary, left_on='Unnamed: 0', right_on='name')
merge1.head()
```

```python
merge1.count()
```

```python
url_3='datasets/tuition_cost.csv'
```

```python
tuition=pd.read_csv(url_3)
tuition.head()
```

```python
merge2=merge1.merge(tuition, left_on='Unnamed: 0', right_on='name')
for x in merge2['Unnamed: 0']:
    print(x)
    
```

```python
merge2.count()
```

```python
merge2.to_csv('merged_data.csv')
```

```python
merge3=costs_college.merge(tuition, left_on='Unnamed: 0', right_on='name')
merge3.head()
```

```python
merge3.name.count()
```

```python
merge4=tuition.merge(salary, left_on='name', right_on='name')
merge4.head()
```

```python
merge4.name.count()
```

```python
merge4_clean=merge4.dropna()
merge4_clean.name.count()
```

```python
merge4.to_csv('tuition_vs_salary.csv')
```

```python
merge4_clean.count()
```

```python
merge4_clean.to_csv('tuition_vs_salary.csv')
```

```python
merge3_clean=merge3.dropna()
merge3_clean.count()
```

```python
merge3_clean.to_csv('costs_college_vs_tuition.csv')
```

```python
rds_connection_string = "postgres:postgres@localhost:5432/project-2"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
merge2.to_sql(name='merged_data', con=engine, if_exists='append', index=False)
```

```python
merge4_clean.to_sql(name='tuition_vs_salary', con=engine, if_exists='append', index=False)
```
